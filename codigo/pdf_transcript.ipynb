```python
import os  # importamos ruta de carpeta a utilizar

ruta_base = "/content/drive/My Drive/NLP_Repo_Sagrada/codigo"
carpeta_entrada = os.path.join(ruta_base, "docx_pdfs")
archivo_salida = "/content/drive/My Drive/NLP_Repo_Sagrada/datos/informacion/pdfs_transcriptos.txt"
```

```python
!pip install pdfplumber

import pdfplumber

def extraer_texto_pdf(ruta_pdf):
    texto = ""
    with pdfplumber.open(ruta_pdf) as pdf:
        for pagina in pdf.pages:
            texto += pagina.extract_text() + "\n"
    return texto
```

```python
!pip install python-docx

from docx import Document

def extraer_texto_docx(ruta_docx):
    doc = Document(ruta_docx)
    return "\n".join([p.text for p in doc.paragraphs])
```

```python
contenido_total = ""

archivos = os.listdir(carpeta_entrada)

for archivo in archivos:
    ruta = os.path.join(carpeta_entrada, archivo)

    if archivo.endswith(".pdf"):
        print(f"Procesando PDF: {archivo}")
        contenido_total += extraer_texto_pdf(ruta) + "\n"

    elif archivo.endswith(".docx"):
        print(f"Procesando DOCX: {archivo}")
        contenido_total += extraer_texto_docx(ruta) + "\n"

# Guardar todo el texto extraído
with open(archivo_salida, "w", encoding="utf-8") as f:
    f.write(contenido_total)

print(f"\n✅ Texto combinado guardado en: {archivo_salida}")
```
